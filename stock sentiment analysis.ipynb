```python
import pandas as pd
```

```python
data=pd.read_csv("Data.csv",encoding= 'ISO-8859-1')
```

```python
data.head()
```

```python
data.tail()
```

```python
train_data = data[data["Date"] < "20150101"]
test_data = data[data["Date"] > "20141231"]
```

```python
train_data['Label']
```

```python
df = train_data.iloc[:,2:27]
```

```python
# Removing punctuations 
df.replace("[^a-zA-Z]"," ",regex=True, inplace=True)
#df=df.replace("[^a-zA-Z]"," ",regex=True)
```

```python
# Renaming column names for ease of access
list=[i for i in range(25)]
new_index=[str(x) for x in list]
df.columns=new_index
df.head()
```

```python
# Converting headlines to lower case
for index in new_index:
    df[index]=df[index].str.lower()
df.head(1)

```

```python
headlines=[]
for i in range(0,len(df.index)):
    headlines.append(" ".join(str(x) for x in df.iloc[i,0:25]))
```

```python
headlines[0]
```

```python
# Implementing Bag of Words
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(ngram_range=(2,2))
traindata=cv.fit_transform(headlines)
```

```python
help(CountVectorizer)
```

```python
# implement RandomForest Classifier
from sklearn.ensemble import RandomForestClassifier
randomclassifier=RandomForestClassifier(n_estimators=200,criterion='entropy')
randomclassifier.fit(traindata,train_data['Label'])
```

```python
#Making predictions for Test data
test_tranform=[]
for i in range(0,len(test_data.index)):
    test_tranform.append(" ".join(str(x) for x in test_data.iloc[i,0:25]))
test_transform_data=cv.transform(test_tranform)
predictions=randomclassifier.predict(test_transform_data)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(test_data['Label'],predictions)
```

```python
# Implementing Tf-Idf Vectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
Tf=TfidfVectorizer(ngram_range=(2,2))
train_tf_data = Tf.fit_transform(headlines)
```

```python
# implement RandomForest Classifier
randomclassifier.fit(train_tf_data,train_data["Label"])
```

```python
## Predict for the Test Dataset
test_tf_data = Tf.transform(test_tranform)
predic = randomclassifier.predict(test_tf_data)
```

```python
accuracy_score(test_data['Label'],predic)
```
